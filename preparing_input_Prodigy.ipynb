```python
short = []
text = []
for index, sent in enumerate(list(df_stripped.sentence)):
    if len(str(sent)) < 60:
        short.append(index)
        text.append(sent)
```

```python
indexes = []
checks = []
keepers = [".", "?", "!", ":"]
for index, text in enumerate(examples):
    text = text.rstrip()
    if text[-1] not in keepers:
        indexes.append(index)
        checks.append(text)
```
